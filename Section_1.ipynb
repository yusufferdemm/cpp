# STD
---
std, C++ dilinde "Standard Template Library" (STL) ve C++ Standard Library'nin (C++ Standart Kütüphanesi) bir parçası olan "standard"ın kısaltmasıdır. C++ dilinin standart kütüphanesi, programcılara yaygın olarak kullanılan veri yapıları, algoritmalar ve işlevler sunar ve C++ programlarının taşınabilirliğini ve yeniden kullanılabilirliğini artırır. "std" adı genellikle C++ programlarında kullanılan C++ Standart Kütüphanesi'nin bileşenlerini temsil eder.

C++ Standart Kütüphanesi (C++ Standard Library) şunları içerir:

1. __STL (Standard Template Library):__ Daha önce bahsedilen, vektörler, listeler, deque'ler, sıralı konteynerlar, algoritmalar, iteratörler ve fonksiyon nesneleri gibi veri yapıları ve algoritmaları içerir. Bu, C++ programlarının genellikle kullanılan veri yapıları ve işlevleri kullanmalarını sağlar.
2. __Input/Output İşlemleri (I/O):__ C++ Standart Kütüphanesi, klavyeden veya dosyalardan giriş almak ve ekrana veya dosyalara çıktı yazmak için kullanılan işlevleri içerir. Bunlar iostream, fstream, ifstream, ofstream gibi başlıklar altında bulunur.
3. __Dize İşlemleri:__ string sınıfı gibi dize işlemleri için işlevleri içerir. Bu işlevler, metin işleme ve manipülasyonu için kullanılır.
4. __Sayı ve Matematik İşlemleri:__ C++ Standart Kütüphanesi, sayılarla işlem yapmayı kolaylaştıran işlevleri içerir. Bu işlevler matematik işlemleri, rastgele sayı üretimi ve zaman işlemleri gibi alanlarda yardımcı olur.
5. __Veri Düzenlemesi:__ Dizi ve çeşitli koleksiyonların düzenlenmesi ve sıralanması için işlevleri içerir. Örneğin, vector sınıfı dizilerin boyutunu otomatik olarak yönetirken, sort işlevi sıralama işlemini gerçekleştirir.
6. __Standart Veri Tipleri:__ Temel veri türleri (int, double, char vb.) ve bu türlerle çalışmayı kolaylaştıran işlevler içerir.
7. __Hata Ayıklama ve İstisna İşleme:__ Hata ayıklama ve istisna işleme işlevleri, hataları tespit etmek ve yönetmek için kullanılır. assert ve try-catch gibi yapılar bu bölüme aittir.


"std" adı, C++ Standart Kütüphanesi'nin bileşenlerini kullanırken belirtmek ve bu işlevlerin ve sınıfların C++ Standart Kütüphanesi'ne ait olduğunu vurgulamak için kullanılır. Örneğin, std::cout, "std" adının bir örneği olarak, C++ Standart Kütüphanesi'nden cout işlevine erişmek için kullanılır.

C++ programlarının başlangıç noktası `main` fonksiyonudur. Bu fonksiyon programın ana giriş noktasıdır ve program çalıştırıldığında otomatik olarak çağrılır. main fonksiyonu şu şekilde tanımlanır:


---
```c++
int main() {
    // Program kodu buraya yazılır
    return 0; // Programın başarıyla sonlandığını belirtmek için genellikle 0 döndürülür
}


```

- return 0; ifadesi, programın başarıyla sonlandığını belirten bir değer döndürür. Programın başarısız olduğu durumlarda farklı bir değer döndürülebilir.

# if / else / else if /switch-case
---
1. `if` İfadesi: Koşulun doğru (true) olduğu durumda belirli bir kod bloğunu çalıştırmak için kullanılır.
2. `if-else` İfadesi: Koşula bağlı olarak iki farklı kod bloğunu kontrol etmek için kullanılır. Koşul doğruysa if bloğu çalışır, aksi takdirde else bloğu çalışır.
3. `else` İfadesi: Bir if veya if-else ifadesinin koşulu yanlış olduğunda çalıştırılacak kod bloğunu belirtmek için kullanılır.
4. `else if` İfadesi: Birden fazla koşulu kontrol etmek için kullanılır. İlk if ifadesinin koşulu yanlışsa, sıradaki else if ifadeleri sırayla kontrol edilir ve ilk doğru koşul bulunduğunda ilgili kod bloğu çalıştırılır.
5. `switch-case` İfadesi: Bir değişkenin değerine göre farklı kod bloklarını çalıştırmak için kullanılır. switch ifadesi içindeki case etiketleri, değişkenin değeri ile karşılaştırılır ve eşleşen case bloğu çalıştırılır.

---
__örnek ; password oluşturma ve sorma__
```cpp
#include <iostream>

using namespace std;

int main()
{   
    // Sifre belirleme 
    int password ;
    
    cout << "Lütfen bir şifre giriniz :" ;
    cin  >> password ;
    cout << " Şifre Belirlendi." << endl;

    // Sifre sorma 
    int passtry ;
    cout << "Lütfen Şifre giriniz :" ;
    cin >> passtry ;

    if (password == passtry)
    {
        cout << "Giriş Başarılı!"<< endl;
    }
    else 
    {
        cout << "Giriş başarısız :(" << endl;
    }

    return 0;
}    

```

# Switch - case
---
`switch:` İfadenin değerini kontrol eden anahtar kelime olarak kullanılır.

`case`: Kontrol edilecek değerleri tanımlamak için kullanılır. İfade, her case etiketi ile karşılaştırılır.

`break`: Bir case bloğunun sonunda kullanılır ve bu ifade bulunduğu case bloğunu terk eder. Eğer break kullanılmazsa, 
case eşleştiği andan itibaren bir sonraki case bloğu da çalıştırılır.

`default`: Eğer hiçbir case etiketi ile eşleşme yoksa, default bloğu çalışır. Bu bloğun kullanımı isteğe bağlıdır.

---
```cpp
#include <iostream>

using namespace std;

int main() {
    int gun;
    cout << "Haftanın kacinci gunu (1-7): ";
    cin >> gun;

    switch (gun) {
        case 1:
            cout << "Pazartesi" << endl;
            break;
        case 2:
            cout << "Salı" << endl;
            break;
        case 3:
            cout << "Çarşamba" << endl;
            break;
        case 4:
            cout << "Perşembe" << endl;
            break;
        case 5:
            cout << "Cuma" << endl;
            break;
        case 6:
            cout << "Cumartesi" << endl;
            break;
        case 7:
            cout << "Pazar" << endl;
            break;
        default:
            cout << "Geçersiz gün numarası!" << endl;
    }

    return 0;
}

```

# continue , break
---

`break` ifadesi, döngüyü aniden sonlandırmak veya bir şart karşılandığında döngüyü terk etmek için kullanılır. Genellikle if ifadesi içindeki bir koşul karşılandığında veya belirli bir işlem yapıldığında döngüden çıkmak için kullanılır. İşte bir örnek:

```c
for (int i = 1; i <= 10; i++) {
    if (i == 5) {
        break; // i değeri 5 olduğunda döngüyü sonlandırır
    }
    cout << i << " ";
}


```


`continue` ifadesi, döngünün mevcut dönemini sonlandırır ve döngü koşulunun kontrolüne geri döner. Yani, continue ifadesi ile karşılaşıldığında döngüdeki geri kalan kodlar atlanır ve bir sonraki döneme geçilir. İşte bir örnek:

```c
for (int i = 1; i <= 10; i++) {
    if (i == 5) {
        continue; // i değeri 5 olduğunda dönemin geri kalanı atlanır
    }
    cout << i << " ";
}

```

Bu örnekte, continue ifadesi, i değeri 5 olduğunda mevcut dönemi atlar ve bir sonraki döneme geçer.

## Function Overloading
---

C++ programlama dilinde "function overloading," aynı isme sahip ancak farklı parametre listelerine sahip birden fazla fonksiyonun tanımlanmasına verilen bir addır. Bu, aynı adı taşıyan fonksiyonların farklı parametre türlerini veya farklı sayıda parametreyi kabul etmesini sağlar. C++'da işlev aşırı yüklemesi yapmak, daha okunabilir ve esnek kod yazmanızı sağlar.


Fonksiyon overloading'in C++ dilindeki özellikleri şunlardır:

- Aynı isim, farklı parametre listeleri: Aynı isme sahip birden fazla işlev tanımlayabilirsiniz, ancak bu işlevlerin parametre listeleri farklı olmalıdır. Parametre sayısı veya parametre türleri farklı olabilir.
- Derleme zamanı çözünürlüğü: C++ derleme zamanında hangi işlemin çağrılacağını belirler. 

İşlev çağrısı, derleyici tarafından en uygun parametre listesine sahip işlevin bulunmasına dayalı olarak çözümlenir.
İşte C++ dilinde fonksiyon aşırı yüklemesine birkaç örnek:

```C
#include <iostream>

// İki tamsayıyı toplayan işlev
int toplama(int a, int b) {
    return a + b;
}

// İki gerçek sayıyı toplayan işlev
double toplama(double a, double b) {
    return a + b;
}

// Üç tamsayıyı toplayan işlev
int toplama(int a, int b, int c) {
    return a + b + c;
}

int main() {
    int sonuc1 = toplama(5, 3);
    double sonuc2 = toplama(2.5, 3.7);
    int sonuc3 = toplama(1, 2, 3);

    std::cout << "Sonuc 1: " << sonuc1 << std::endl;
    std::cout << "Sonuc 2: " << sonuc2 << std::endl;
    std::cout << "Sonuc 3: " << sonuc3 << std::endl;

    return 0;
}

```

2. __static Anahtar Kelimesi:__

static anahtar kelimesi, bir değişkenin, bir sınıfa veya bir işlev içinde yerel olmayan bir değişken olarak tanımlandığını belirtmek için kullanılır. static anahtar kelimesi kullanıldığında, değişken programın çalışması süresince yaşamını sürdürür ve programın çalışmasından bağımsızdır. Örneğin:

```cpp 
class Counter {
private:
    static int count;  // static üye değişken

public:
    Counter() {
        count++;
    }

    static int getCount() {
        return count;
    }
};

int Counter::count = 0;  // static üye değişkenin tanımı

```
---
mesela ben aşağıdaki programı çalıştırdığımda alt alta , 1 2 ve 3 yazıyor .

```C
#include <iostream>
#include <chrono>


using namespace std;

void fonksiyon(){
    static int sayac =0;
    sayac++;
    cout << sayac << endl ;
}

int main()
{
    fonksiyon();
    fonksiyon();
    fonksiyon();

    return 0;
}



```

# Call by Value/Reference
---

C++ dilinde, fonksiyonlara parametre olarak değerlerin nasıl geçirileceğini belirtmek için "call by value" ve "call by reference" terimleri kullanılır.

__Call by Value (Değer ile Çağırma):__

Fonksiyonlara parametre olarak değerlerin geçirilmesi anlamına gelir. Bu durumda, fonksiyon, parametre olarak geçirilen değerin bir kopyasını alır ve bu kopya üzerinde işlem yapar. Bu nedenle, fonksiyon içinde yapılan değişiklikler, orijinal değeri etkilemez.
Örneğin:

```cpp
void increment(int x) {
    x++;
}

int main() {
    int num = 10;
    increment(num);
    cout << "num: " << num;  // num: 10
    return 0;
}

```
__Call by Reference (Referans ile Çağırma):__

Fonksiyonlara parametre olarak değerlerin referanslarının geçirilmesi anlamına gelir. Bu durumda, fonksiyon, orijinal değişkenin bellek adresini alır ve bu adres üzerinden çalışır. Yapılan değişiklikler orijinal değişkeni etkiler.
Örneğin:

```cpp
void incrementByReference(int& x) {
    x++;
}

int main() {
    int num = 10;
    incrementByReference(num);
    cout << "num: " << num;  // num: 11
    return 0;
}

```
